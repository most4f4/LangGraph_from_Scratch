# Exercise 1: Create a StateGraph with single input

```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class UserInput(TypedDict):
    name: str

def complement_user_input(state: UserInput) -> UserInput:
    state['name'] = state['name'] + ", you're doing an amazing job learning LangGraph!"
    return state
```

```python
graph = StateGraph(UserInput)

graph.add_node("complement", complement_user_input)

graph.set_entry_point("complement")
graph.set_finish_point("complement")

app = graph.compile()
```

```python
from IPython.display import display, Image

display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"name": "Bob"})
```
